## Check correlation

#### Imports

```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import csv
```

#### Read data

```python
data = pd.read_csv('istherecorrelation.csv', sep=';')
```

```python
year = list(data['Year'])
wo = list(data['WO [x1000]'])
consumption = list(data['NL Beer consumption [x1000 hectoliter]'])


for i in range(len(wo)):
    wo[i] =  float(wo[i].replace(',','.'))
```

#### Plot

```python
fig = plt.plot()

plt.scatter(year, wo, color = 'red', marker = 'o',label = 'WO [x1000]')
plt.ylabel('WO [x1000]', color = 'red', fontsize = 14)
plt.xlabel('Year', fontsize = 16)

plt.twinx()
plt.scatter(year, consumption, color = 'blue', marker = 'o', label = 'Beer Consumption\n(x1000 hectoliter)')
plt.ylabel('Beer Consumption\n(x1000 hectoliter)', color = 'blue', fontsize = 14)

plt.savefig('correlation_1.pdf', dpi=300)
```

```python
fig = plt.plot()

plt.scatter(wo, consumption, color = 'black', marker = 'o',label = 'WO [x1000]')
plt.xlabel('WO [x1000]', fontsize = 14)
plt.ylabel('Beer Consumption\n(x1000 hectoliter)', fontsize = 14)

plt.savefig('correlation_2.pdf', dpi=300)
```
